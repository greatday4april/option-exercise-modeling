```python
from tax import get_fy_projection, Event, FYS, FMV_AT_EXERCISE
import pandas as pd
```

### FY2022 and FY2023 Planning

```python
first = Event(6377, 17, "exercise", "iso")
events = [
    first,
    Event(20500, 21, "sale", "nso", FMV_AT_EXERCISE),
    Event(6377, 15+12, "sale", "iso", first.price)
]

pd.DataFrame([
    get_fy_projection(True, FYS[2], events[0:2]),
    get_fy_projection(False, FYS[2], events[0:2]),
    get_fy_projection(True, FYS[3], events[2:]),
    get_fy_projection(False, FYS[3], events[2:])
])
```

### FY2024

```python
first = Event(3000, 20, "exercise", "iso")
second = Event(3377, 27, "exercise", "iso")

nso_exercise = Event(20000, 40, "exercise", "nso")
nso_sell = Event(20000, 40, "sale", "nso", nso_exercise.price)

events = [
    Event(10500, 18, "sale", "nso", FMV_AT_EXERCISE),
    first,
    second,
    Event(10000, 18 + 9, "sale", "nso", FMV_AT_EXERCISE),
    Event(3000, 20+12, "sale", "iso", first.price),
    Event(3377, 27+12, "sale", "iso", FMV_AT_EXERCISE),
    nso_exercise,
    nso_sell,
    
]

fy2022_events = [e for e in events if e.month < 21]
fy2023_events = [e for e in events if e.month >= 21 and e.month < 33]
fy2024_events = [e for e in events if e.month >= 33]

pd.DataFrame([
    get_fy_projection(True, FYS[2], fy2022_events),
    get_fy_projection(False, FYS[2], fy2022_events),
    get_fy_projection(True, FYS[3], fy2023_events),
    get_fy_projection(False, FYS[3], fy2023_events),
    get_fy_projection(True, FYS[4], fy2024_events),
    get_fy_projection(False, FYS[4], fy2024_events)
])
```
