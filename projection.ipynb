```python
from tax import get_fy_projection, Event, FYS, FMV_AT_EXERCISE
import pandas as pd
```

### FY2022 and FY2023 Planning

```python
first = Event(6377, 15, "exercise", "iso")
events = [
    first,
    Event(20500, 18, "sale", "nso", FMV_AT_EXERCISE),
    Event(6377, 15+12, "sale", "iso", first.price)
]

pd.DataFrame([
    get_fy_projection(True, FYS[2], events[0:2]),
    get_fy_projection(False, FYS[2], events[0:2]),
    get_fy_projection(True, FYS[3], events[2:]),
    get_fy_projection(False, FYS[3], events[2:])
])
```

*TO minimize the AMT TAX, try moving some ISO exercise to 2023*

```python
first = Event(3000, 15, "exercise", "iso")
second = Event(3377, 27, "exercise", "iso")
events = [
    first,
    Event(20500, 18, "sale", "nso", FMV_AT_EXERCISE),
    second,
    Event(3000, 15+12, "sale", "iso", first.price)
]

pd.DataFrame([
    get_fy_projection(True, FYS[2], events[0:2]),
    get_fy_projection(False, FYS[2], events[0:2]),
    get_fy_projection(True, FYS[3], events[2:]),
    get_fy_projection(False, FYS[3], events[2:])
])
```
